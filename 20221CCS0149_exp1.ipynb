```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,40,50]
name=["Jan", "Feb", "March", "April"]
Color=["Pink","SkyBlue","Yellow","Silver"]
value=[0.2,0,0.5,0]
plt.pie(x, labels=name, colors=Color,explode=value,shadow=True)
plt.show()
```

```python
 import numpy as np
X=np.array(["A","B","C","D"])
Y=np.array([3,8,1,10])
Color=["Green","Blue","Black","Brown"]
plt.bar(X,Y,color=Color)
plt.show()
```

```python
import numpy as np
X=np.array(["A","B","C","D"])
Y=np.array([2,4,6,8])
Color=["Green","Blue","Black","Brown"]
Size=[900,500,600,400]
plt.scatter(X,Y,color=Color,marker="*",edgecolor="Green",s=Size,alpha=0.8)
plt.title("scatter",fontsize=20)
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.show()
```
